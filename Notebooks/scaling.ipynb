```python
import os, ast
import pandas as pd
import matplotlib.pyplot as plt

path_1 ='../timer/' 
```

```python
names = sorted(os.listdir(path_1)) 

base = pd.read_csv(path_1+"log_1.csv", delimiter=',', header=0)
base_time = base.training_time[0]
#print(base_time)

efficiencys = []
efficiencys.append((1, 100))
for i,name in enumerate(names):
    path = path_1 +'/'+names[i]
    Df = pd.read_csv(path, delimiter=',', header=0)
    if names[i] != "log_1.csv":
        training_time = Df.training_time[0]
        efficiency = base_time / training_time * 100
        efficiencys.append((int(name[4:-4]),efficiency))

#print(efficiencys)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5))
for i,name in enumerate(names):
    
    path = path_1 +'/'+names[i]
    Df = pd.read_csv(path, delimiter=',', header=0)
   
    ax1.bar(int(name[4:-4]), Df.training_time[0], label=name)
    ax2.bar(efficiencys[i][0], efficiencys[i][1], label=name)

ax1.legend()
ax2.legend()

ax1.set_ylabel("total training time [sec]")
ax2.set_ylabel("efficiency [%]")

ax1.set_xlabel("MPI ranks")
ax2.set_xlabel("MPI ranks")

```
